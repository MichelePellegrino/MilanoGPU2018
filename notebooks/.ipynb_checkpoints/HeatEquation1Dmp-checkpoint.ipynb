```python
# CPU
class HeatEquation1D(object):
    """
        This class approx. heat equation, expl. finite diff. (DOCUMENTATION)
    """
    
    def __init__(self, u0, kappa, dx, dt):
        self.u0 = u0.copy() # NB mutable/non-mutable
        self.u1 = np.empty_like(self.u0)
        self.kappa = kappa
        self.dx = dx
        self.dt = dt
        self.nx = self.u0.shape[0]-2
    
    def step(self):
        for i in range(1, self.nx+1):
            self.u1[i] = self.u0[i] - \
                - self.kappa*self.dt/self.dx**2 * (self.u0[i-1] - 2.0*self.u0[i] + self.u0[i+1])
        self.u1[0] = self.u1[1]
        self.u1[-1] = self.u1[-2]
        
        # Swap u0, u1
        self.u0, self.u1 = self.u1, self.u0
            
    def download(self):
        return self.u1.copy()
```

```python
# CPU
u0 = np.zeros(17)
u0[8] = 1.0
kappa = 1.0
dx = 1.0
dt = 0.4*(dx**2)/(2*kappa)
cpu_simulator = HeatEquation1D(u0, kappa, dx, dt)
for i in range(15):
    cpu_simulator.step()
    result = cpu_simulator.download()
    # print(result)
    fig = plt.figure()
    plt.plot(result)
```

```python
# GPU
u0 = np.zeros(5, dtype=np.float32)
u0[2] = 1.0
kappa = 1.0
dx = 1.0
dt = 0.4*(dx**2)/(2*kappa)
gpu_simulator = HeatEquation1DGPU()
gpu_simulator.initialize(u0, kappa, dx, dt)
```

```python
clean_tests()

def test_HeatEquation1D():
    u0 = np.zeros(15)
    kappa = 1.0
    dx = 1.0
    dt = 0.4*(dx**2)/(2*kappa)
    simulator = HeatEquation1D(u0, kappa, dx, dt)
    
    assert np.all( u0 == simulator.u0 )
    assert simulator.u0.shape[0] == simulator.u1.shape[0]
    assert kappa == simulator.kappa
    assert dx == simulator.dx
    assert dt == simulator.dt
    
    u0 = np.zeros(7)
    u0[3] = 1.0
    kappa = 1.0
    dx = 1.0
    dt = 0.4*(dx**2)/(2*kappa)
    simulator = HeatEquation1D(u0, kappa, dx, dt)
    for i in range(10):
        simulator.step()
        result = simulator.download()
        print(result)
        assert np.sum(result[1:-1]) == pytest.approx(1.0)
        assert result[0] == result[1]
        assert result[-1] == result[-2]
    
run_pytest(filename='HeatEquation1Dmp.ipynb', pytest_options=['-vvv'])
```
